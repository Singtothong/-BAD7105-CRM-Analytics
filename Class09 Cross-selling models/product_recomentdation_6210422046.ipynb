```python
# ลบ Playsation5 กับ PS5 เป็นอันเดียวกัน
df = df.drop('PS5', axis = 1)
# ลบ timestamp
df = df.drop('Timestamp', axis = 1)
```

```python
df.replace({'ไม่เคยซื้อ': 0, 'เคยซื้อ': 1}).head()
# keyword ไม่ตรงกัน
```

```python
one_hot_df = df.replace(regex={r'^ไม่': 0, 'เคย': 1})
one_hot_df
```

```python
df['Logitech Mx Master 3 Mouse']
```

```python
one_hot_df['Logitech Mx Master 3 Mouse']
```

```python
transaction_df =  one_hot_df.unstack().reset_index()
```

```python
transaction_df = transaction_df[transaction_df[0] != 0]
```

```python
transaction_df = transaction_df[['level_0','level_1']].rename({'level_0': 'Item', 'level_1': 'Transaction'}, axis=1)
```

```python
import matplotlib as mpl
mpl.font_manager.fontManager.addfont('THSarabunChula-Regular.ttf')
mpl.rc('font', family='TH Sarabun Chula')
```

```python
import numpy as np
import matplotlib.pyplot as plt

df_for_top10_Items=transaction_df['Item'].value_counts().head(10)
Item_array= np.arange(len(df_for_top10_Items))

plt.figure(figsize=(15,5))
Items_name=list(df_for_top10_Items.keys())
plt.bar(Item_array,df_for_top10_Items.iloc[:])
plt.xticks(Item_array,Items_name)
plt.title('Top 10 most selling items')
plt.show()
```

```python
hot_encoded_df=transaction_df.groupby(['Transaction','Item'])['Item'].count().unstack().reset_index().fillna(0).set_index('Transaction')
```

```python
hot_encoded_df.head(5)
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
hot_encoded_df = hot_encoded_df.applymap(encode_units)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(hot_encoded_df, min_support=0.01, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(10)
```

```python
rules[ (rules['lift'] >= 1) &
       (rules['confidence'] >= 0.5) ]
```
